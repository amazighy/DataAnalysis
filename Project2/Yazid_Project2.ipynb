## deleting the features that have the same values for all the observation
To find the id or the features that have unique values for all the observations, I wrote a function that finds information about the unique values as well as the percentages of the unique values. 
> The function takes 3 argument 
- **a datafrme**
- **unique**: it permits to view features with a certain percentage of unique values.
- **action** it permits to either print info or returns the of features with the indicated percentage of unique values


```python
def IDs_Finder(df,unique=100,action = 'print'):
    feature = []
    number = []
    for feat in df.columns:
        lenght=len(df[feat].unique())
        x = (len(df[feat].unique())/len(df))*100
        if x >= unique:
            feature.append(feat)
            number.append(lenght)
        d = dict(zip(feature, number))
    if action == "return":
        return list(d.keys())
    elif len(feature) == 0:
        return []
    else:
        print('{:30.25}'.format('The feature'), '{:30.20}'.format('# unique values'),'{:30.20}'.format('% unique values'))
        print("------------------------------------------------------------------------------")

        for feature, value in sorted(d.items(), key=lambda item: item[1], reverse=True):
            per = value / (len(df)) * 100 
            print(
                "{:33.20}".format(feature),
                "{:33.20}".format(str(value)),
                "{:4.5}".format(str(per)),
                "{:4.6}".format("%"),
            ) 
```

>>#### from the above figures, we can see that most of the variables are extremely skewed. 
below I'm writing a function that permits us to visualize the distribution of variables before and after the **Yeo-Johnson Transformation**. The function plots side by side a **histogram** of the original variable and the transformed variable, as well as their  **Q-Q plots**.
- motivation for using Yeo-Johnson is [here](https://www.stat.umn.edu/arc/yjpower.pdf):

```python
import scipy.stats as stats
def diagnostic_plots(df, variable): 
    df=df.copy()
    print('')
    print("    ---------------------------------------------------------------------------------------------------------------")
    print("    ******************************Visulaizing the transformation of",variable,"************************************")
    print("    ---------------------------------------------------------------------------------------------------------------")
    df[variable] = df[variable].astype('float')
    df['GrLivArea_yeojohnson'], param = stats.yeojohnson(df[variable]) 
    plt.figure(figsize=(15,8))
    plt.subplot(2, 2, 1)
    plt.gca().set_title("original distrubution of "+variable)
    df[variable].hist(bins=30)
    plt.subplot(2, 2, 2)
    plt.gca().set_title("distrubution of "+variable+" after Yeo-Johnson" )
    df['GrLivArea_yeojohnson'].hist(bins=30)
    plt.subplot(2, 2, 3)
    plt.gca().set_title("original Q-Q plot of "+variable)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.subplot(2, 2, 4)
    stats.probplot(df['GrLivArea_yeojohnson'], dist="norm", plot=plt)
    plt.gca().set_title(" Q-Q plot of "+variable+" after Yeo-Johnson" )
    plt.show() 
```

```python
# looping over the all the numeric varibles and plotting the transformations
for feat in data.select_dtypes(exclude=['object']).columns:   
    diagnostic_plots(data, feat)
```

> ## Mean, Median, and Mode Imputation
To figure out the best way to impute the numeric variables, I wrote a function that creates 3 new columns in the dataframe for each feature with missing data.  The new features are named with the name of the variable with missing data **+_** the imputation method. For example, **feature_mean**  is the original features with the missing data imputed with the mean. The function imputes the missing data with the mean, median, and mode. 
The **mean_median_mode_imputation**  function takes a dataframe and a list of missing data. It returns the variance, mean, and median for the original features and the newly engineered features. This is done to compare the mean, median of the engineered features to figure out which one preserves the distribution of the original features.
To visualize the results the function plots the distribution and the box plots of all features to visualize the differences.


```python
def mean_median_mode_imputation(df, Features_NaN): 
    df=df.copy()
    print('{:25.25}'.format('The Feature'), '{:25.20}'.format('The Variance'),'{:25.20}'.format('The Mean'),'{:25.20}'.format('The Median'),'{:25.20}'.format('The Mode'))
    print("--------------------------------------------------------------------------------------------------------------------") 
    for feat in Features_NaN:
        df[feat+'_median']=df[feat].fillna(df[feat].median())
        df[feat+'_mean']=df[feat].fillna(df[feat].mean()) 
        df[feat+'_mode']=df[feat].fillna(df[feat].mode().values[0]) 
        print('{:25.25}'.format(feat), '{:25.20}'.format(str( np.round(df[feat].var(),2) )),'{:25.20}'.format(str(np.round(df[feat].mean(),2))),'{:25.20}'.format(str(np.round(df[feat].median(),2))),'{:25.20}'.format(str(np.round(df[feat].mode().values[0],2))))
        print('{:25.25}'.format(feat+'_mean'), '{:25.20}'.format(str( np.round(df[feat+'_mean'].var(),2) )),'{:25.20}'.format(str(np.round(df[feat+'_mean'].mean(),2))),'{:25.20}'.format(str(np.round(df[feat+'_mean'].median(),2))),'{:25.20}'.format(str(np.round(df[feat+'_mean'].mode().values[0],2))))
        print('{:25.25}'.format(feat+'_mode'), '{:25.20}'.format(str( np.round(df[feat+'_mode'].var(),2) )),'{:25.20}'.format(str(np.round(df[feat+'_mode'].mean(),2))),'{:25.20}'.format(str(np.round(df[feat+'_mode'].median(),2))),'{:25.20}'.format(str(np.round(df[feat+'_mode'].mode().values[0],2))))
        print('{:25.25}'.format(feat+'_median'), '{:25.20}'.format(str( np.round(df[feat+'_median'].var(),2) )),'{:25.20}'.format(str(np.round(df[feat+'_median'].mean(),2))),'{:25.20}'.format(str(np.round(df[feat+'_median'].median(),2))),'{:25.20}'.format(str(np.round(df[feat+'_median'].mode().values[0],2))))
        print('')
        
        fig = plt.figure(figsize=(15,6))
        ax = fig.add_subplot(111)  
        df[feat].plot(kind='kde', ax=ax)
        df[feat+'_median'].plot(kind='kde', ax=ax, color='red')
        df[feat+'_mode'].plot(kind='kde', ax=ax, color='green')
        df[feat+'_mean'].plot(kind='kde', ax=ax, color='black')
        lines, labels = ax.get_legend_handles_labels()
        ax.legend(lines, labels)
        plt.title('Distrubutions for '+feat)

        figure = plt.figure(figsize=(15,6))
        boxplot = df.boxplot(column=[feat, feat+'_mode', feat+'_mean',feat+'_median'],figsize=(15,6))
        plt.title('Box plot for '+feat)
        plt.ylabel('Density')
        
        
```

>From analyzing the results of the **mean_median_mode_imputation** function, I think the best way to impute features with missing data is as follows:
- I will impute **GarageYrBlt** with **median** as it best preserves the distribution of the missing data.
- I will impute **MasVnrArea ** and** LotFrontage with **mean** as it best preserves the distribution of the missing data.
To perform the imputation I wrote afunction that uses the Feature engine module.<br>
 The **mean_median_imputer** takes 4 arguments: 
- **The train_set, The test_set, The list of features** we want to impute, and **The imputation method** 
- The function makes an instance of the **missing_data_imputers** imported from **feature_engine**, then **fit it** to the train_set and returns the **transformed** train and test set 


```python
def mean_median_imputer(train_set, test_set, features, method):
    from feature_engine import missing_data_imputers as mdi
    imputer= mdi.MeanMedianImputer(imputation_method=method, variables=features)
    imputer.fit(train_set)
    return imputer.transform(train_set), imputer.transform(test_set)
```

To replace the features that have labels that are present in less than 1 percent of the observations, I wrote the function below.
> The function takes 5 arguments: 
- **the train_set**, 
- **test_set**, 
- **rare_list**: list of features that have  labels we want to replace by **rare**
- **target**: the feature wee want to predict 
- **rare_perc**: to manipulate the percentage, meaning we might want to replace by rare the labels that occur in less than 2 percent instead.
> The function transforms the train and test set to account for the new labels with rare.


```python
def replacing_with_rare(train_set,test_set, rare_list, target, rare_perc): 
    for var in rare_list:
        df_copy = train_set.groupby(var)[target].count() / len(train_set)
        train_set[var] = np.where(train_set[var].isin(df_copy [df_copy  > rare_perc].index), train_set[var], 'rare')
        test_set[var] = np.where(test_set[var].isin(df_copy [df_copy  > rare_perc].index), test_set[var], 'rare')
replacing_with_rare(train_set, test_set, cardinality_3,'SalePrice', 0.01 )
```

>> ###  Ordinal encoding
To perform ordinal encoding I wrote a function that replaces the **labels** of each feature with an **integer**.
The **integer value** corresponds to the mean of each label’s house sales. in other words the highest the mean of the category the hight the integer it is assigned. This an attempt to create a monotonic relationship with the target.


```python
def mean_ordinal_encoder(train_set, test_set, feat_list, target):
    for var in feat_list:
        ordered_labels = train_set.groupby([var])[target].mean().sort_values().index
        ordinal_label = {k: i for i, k in enumerate(ordered_labels, 0)}
        train_set[var] = train_set[var].map(ordinal_label)
        test_set[var] = test_set[var].map(ordinal_label)
```

>> ### 1) Get the correlated features.
To get the correlated features, I wrote a function that takes 3 arguments: a DataFrame, a threshold and a third argument to choose what method of correlation we are interested in. If we set  **Brute_force=True** the function will return a list of features that are corelated with at least another features. Else If, we want to investigate further the corelated features we set **Brute_force=False** the function will return a DataFrame of all pairs of correlated features.

```python
def correlation(df, threshold,Brute_force=False):
    # Set of all the names of correlated columns
    column1 = []
    column2 = []
    absolu  = []
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
             #evaluate all possilbe pairs of features and 
            # compare the absolute value against the the threshold
            if abs(corr_matrix.iloc[i, j]) > threshold:
                # getting the name of column
                col1 = corr_matrix.columns[i] 
                col2 = corr_matrix.columns[j]
                abso = abs(corr_matrix.iloc[i, j])
                absolu.append( abso)
                column1.append(col1)
                column2.append(col2)
                
    if Brute_force is True:
        return list(set(column1))
    else:
        df = pd.DataFrame(list(zip(column1,column2,absolu)), 
               columns =['feature1', 'feature2','correlation']) 
    return df
```

```python
# returning correlated features in a dataframe 
corr_df=correlation(XX_train, 0.75)
corr_df
```

From the above DataFrame we can see that there some features that are correlated with more than one feature. Therefore depending on how many features we want to end up  with in the DataFrame, we can keep only one feature from each group.

>> #### 1-a)  getting all the correlated groups
Below I write a function that compares each feature with all of the other features it correlated with. This function takes as an argument the resultant DataFrame from the above correlation function and it returns all the groups of all correlated features.

```python
def correlated_groups(df):
    feature1 = []
    groups = []
    for feature in df.feature1.unique():
        if feature not in feature1:
            # find all features correlated to a single feature
            correlated_block = df[df.feature1 == feature]
            feature1 = feature1 + list(correlated_block.feature2.unique()) + [feature]
            # append the block of features to the list
            groups.append(correlated_block)   
    print('Found {} correlated groups'.format(len(groups)))
    print('out of {} total features'.format(XX_train.shape[1]))          
    return groups
```

```python
# Here i call the correlated_groups and pass it the corelation dataframe
# lets vizualize one group
groups=correlated_groups(corr_df)

groups[3]
```

>> #### 2-a) keep the features with the most predictive power
Alternatively, we could build a machine learning algorithm using all the features from the above groups, and select the the features that have the most predictive. To do this, i wrote the function below. The funstion takes the train and test set and the groups of correlated features from above. To detemine which feature have the most predictive power, i train a random forest model and i use it's **importances_ parameter** to determine the feature with the most predivtive power from each group.

```python
def rf_feat_importance_(XX_train,y_train,groups):

    from sklearn.ensemble import RandomForestRegressor
    
    feat_keep=set()# initialize a list of features to keep
    feat_total=[]# initialize a list of of all corelated features 
    i=1
    # looping over all the groups
    for group in groups:
        feature1=list(group.feature2.unique()) # get the features in the secound column of the group
        feature2=list(group.feature1.unique()) #get the features in the first column of the group
        features=feature1+feature2 # join the features and make a list
        feat_total.append(features) # append the features to the total list
        rf = RandomForestRegressor(n_estimators=200, max_depth=4)# make instance of random forest regressor
        rf.fit(XX_train[features], y_train.values.ravel())# train the model
        print("")
        print("********group",[i],"features importance*******")
        print("----------------------------------------------")
        #*********using rf.feature_importances_ parameter of the ramdom forest
        # get the importance of the features and save it into a series
        importance = pd.concat(# make a serries to capture each feature and its importance
        [pd.Series(features),
         pd.Series(rf.feature_importances_)], axis=1)
        importance.columns = ['feature', 'importance']
        series=importance.sort_values(by='importance',ascending=False)
        feat_keep.add(series['feature'][0])
        print(series)
        i=i+1

    feat_total=sum(feat_total,[]) 
    list_drop=[feat for feat in feat_total if feat not in feat_keep]
    print("")
    print("----------------------------------------------")
    print("----------------------------------------------")
    print("********List of features to keep *************")
    print("----------------------------------------------") 
    print("")
    print(feat_keep)
    print("")
    print("----------------------------------------------")
    print("********List of features drop ****************")
    print("----------------------------------------------") 

    return list(set(list_drop))
```

```python
# here i call the function and acess each feature importance with  random forest **importance_ parameter 
# saving the list to drop to  variable
list_to_drop=rf_feat_importance_(XX_train,y_train,groups )
```

```python
# i am going to delete thse features

XX_train.drop(list_to_drop ,inplace=True, axis=1)
XX_test.drop(list_to_drop, inplace=True, axis=1)
```

### Writing a function to evaluate model performance

```python

def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    return accuracy
```
